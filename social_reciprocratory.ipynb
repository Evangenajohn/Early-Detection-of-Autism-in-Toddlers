```python
import cv2
import mediapipe as mp
import numpy as np
import matplotlib.pyplot as plt
```

```python
mp_face_mesh = mp.solutions.face_mesh
```

```python
# Function to calculate head turn based on nose tip relative to face center
def calculate_head_turn(landmarks):
    # Nose tip index in face mesh: 1
    nose = landmarks[1]
    left_cheek = landmarks[234]
    right_cheek = landmarks[454]

    # Horizontal position relative to cheeks (normalized)
    left_x = left_cheek.x
    right_x = right_cheek.x
    nose_x = nose.x

    # Ratio to measure how much nose deviates from center between cheeks
    center = (left_x + right_x) / 2
    deviation = nose_x - center
    return deviation
```

```python
# Simple smile detection heuristic based on mouth landmarks distance
def detect_smile(landmarks):
    # Mouth corner indices: left 61, right 291
    left_mouth = landmarks[61]
    right_mouth = landmarks[291]
    top_lip = landmarks[13]
    bottom_lip = landmarks[14]

    mouth_width = np.linalg.norm(np.array([left_mouth.x, left_mouth.y]) - np.array([right_mouth.x, right_mouth.y]))
    mouth_height = np.linalg.norm(np.array([top_lip.x, top_lip.y]) - np.array([bottom_lip.x, bottom_lip.y]))

    smile_ratio = mouth_width / mouth_height if mouth_height > 0 else 0
    # Empirically, smile_ratio > 2.0 could indicate smile, adjust threshold as needed
    return smile_ratio > 3.0
```

```python
def analyze_social_reciprocity(video_path):
    cap = cv2.VideoCapture(video_path)
    face_mesh = mp_face_mesh.FaceMesh(static_image_mode=False, max_num_faces=1)

    response_count = 0
    frame_count = 0
    smiling_frames = 0
    head_turn_threshold = 0.02  # tune this threshold for head turn detection

    prev_head_turn = 0
    head_turn_responses = 0

    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break

        frame_count += 1
        rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        results = face_mesh.process(rgb)

        if results.multi_face_landmarks:
            landmarks = results.multi_face_landmarks[0].landmark

            # Head turn detection
            head_turn = calculate_head_turn(landmarks)

            # Detect if head turned noticeably compared to previous frame
            if abs(head_turn - prev_head_turn) > head_turn_threshold:
                head_turn_responses += 1
            prev_head_turn = head_turn

            # Smile detection
            if detect_smile(landmarks):
                smiling_frames += 1

    cap.release()
    face_mesh.close()

    # Rule-based social reciprocity score
    # Number of head turn responses + smiling frames normalized by frame count
    social_response_score = (head_turn_responses + smiling_frames) / frame_count

    if social_response_score < 0.05:
        social_reciprocity = "Low"
    elif social_response_score < 0.15:
        social_reciprocity = "Medium"
    else:
        social_reciprocity = "High"

    return {
        "social_response_score": social_response_score,
        "social_reciprocity_level": social_reciprocity,
        "total_frames": frame_count,
        "head_turn_responses": head_turn_responses,
        "smiling_frames": smiling_frames
    }
```

```python
# Run on your sample video
video_file_path = '/content/drive/MyDrive/videos_task/gesture_delay1.mp4'   # Update path here
result = analyze_social_reciprocity(video_file_path)

print("Social Reciprocity Analysis Result:")
print(result)
```
